### APPLICATION DE APRIORI A WINE

```python



# 1. Préparation des données pour Apriori
print("PRÉPARATION DES DONNÉES POUR APRIORI")
print("="*50)

# Sélectionner les variables à discrétiser pour l'analyse associative
variables = ['alcohol', 'malic_acid', 'ash', 'magnesium', 'total_phenols', 'flavanoids']

# Discrétiser chaque variable en 3 catégories (faible, moyen, élevé)
df_apriori = df[['target']].copy()  # Commencer avec la target

for var in variables:
    # Utiliser la discrétisation par fréquence égale
    df_apriori[f'{var}_cat'] = pd.qcut(df[var], q=3, labels=[f'{var}_faible', f'{var}_moyen', f'{var}_élevé'])

# Ajouter aussi la variable target discrétisée
df_apriori['target_cat'] = df['target'].map({0: 'classe_0', 1: 'classe_1', 2: 'classe_2'})

print("Données après discrétisation :")
print(df_apriori.head())
print(f"\nShape: {df_apriori.shape}")

# 2. Conversion en format transactions
print("\n" + "="*50)
print("CONVERSION EN FORMAT TRANSACTIONS")
print("="*50)

# Créer les transactions (chaque ligne = un vin avec ses caractéristiques)
transactions = []
for i in range(len(df_apriori)):
    transaction = []
    for col in df_apriori.columns:
        if col != 'target':  # Exclure la target originale
            transaction.append(df_apriori.iloc[i][col])
    transactions.append(transaction)

print(f"Nombre de transactions: {len(transactions)}")
print("\nExemple de transaction (premier vin):")
print(transactions[0])

# 3. Encodage one-hot pour Apriori
print("\n" + "="*50)
print("ENCODAGE ONE-HOT")
print("="*50)

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

print(f"Shape encodé: {df_encoded.shape}")
print(f"Nombre d'items: {len(te.columns_)}")
print("\nColonnes (items):")
print(te.columns_)

print(f"Shape encodé: {df_encoded.shape}")
print(f"→ {df_encoded.shape[0]} transactions (vins)")
print(f"→ {df_encoded.shape[1]} items (caractéristiques)")
print(f"Nombre d'items uniques: {len(te.columns_)}")

# Afficher les premiers items
print("\nPremiers 20 items (colonnes):")
for i, item in enumerate(te.columns_[:20]):
    print(f"  {i+1:2d}. {item}")

# Afficher un aperçu de la matrice one-hot
print(f"\nAperçu de la matrice one-hot (5 premières lignes, 10 premières colonnes):")
print(df_encoded.iloc[:5, :10])

# Calculer le support de chaque item (fréquence)
print("\n" + "="*50)
print("SUPPORT DE CHAQUE ITEM (FRÉQUENCE)")
print("="*50)

support_items = df_encoded.mean().sort_values(ascending=False)
print("Top 20 items par support (fréquence):")
for i, (item, support) in enumerate(support_items.head(20).items()):
    print(f"  {i+1:2d}. {item}: {support:.3f} ({support*100:.1f}%)")


# 4. Application de l'algorithme Apriori
print("\n" + "="*50)
print("APPLICATION APRIORI")
print("="*50)

# Trouver les itemsets fréquents avec support minimum
min_support = 0.2  # 20% des transactions
frequent_itemsets = apriori(df_encoded, min_support=min_support, use_colnames=True)

print(f"Itemsets fréquents (support ≥ {min_support}): {len(frequent_itemsets)}")
print("\nTop 10 itemsets les plus fréquents:")
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
print(frequent_itemsets.sort_values('support', ascending=False).head(10))

# 5. Génération des règles d'association
print("\n" + "="*50)
print("RÈGLES D'ASSOCIATION")
print("="*50)

# Générer les règles avec différents seuils
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

print(f"Nombre de règles générées: {len(rules)}")

if len(rules) > 0:
    print("\nTop 10 règles par confidence:")
    rules_sorted = rules.sort_values('confidence', ascending=False).head(10)
    
    # Afficher les règles de manière lisible
    for idx, rule in rules_sorted.iterrows():
        antecedents = list(rule['antecedents'])
        consequents = list(rule['consequents'])
        print(f"Règle: {antecedents} → {consequents}")
        print(f"  Support: {rule['support']:.3f}, Confidence: {rule['confidence']:.3f}, Lift: {rule['lift']:.3f}")
        print()
else:
    print("Aucune règle trouvée avec les paramètres actuels.")

# 6. Analyse spécifique par classe de vin
print("\n" + "="*50)
print("ANALYSE PAR CLASSE DE VIN")
print("="*50)

# Analyser les règles pour chaque classe
for classe in ['classe_0', 'classe_1', 'classe_2']:
    print(f"\n--- RÈGLES POUR LA {classe.upper()} ---")
    
    # Filtrer les règles où la conséquence est la classe
    class_rules = rules[rules['consequents'].apply(lambda x: classe in x)]
    
    if len(class_rules) > 0:
        print(f"Nombre de règles: {len(class_rules)}")
        for idx, rule in class_rules.head(5).iterrows():  # Top 5 par lift
            antecedents = list(rule['antecedents'])
            consequents = list(rule['consequents'])
            print(f"Si {antecedents} alors {consequents}")
            print(f"  Confiance: {rule['confidence']:.3f}, Lift: {rule['lift']:.3f}")
    else:
        print(f"Aucune règle spécifique pour {classe}")

# 7. Visualisation des résultats
print("\n" + "="*50)
print("VISUALISATION DES RÈGLES")
print("="*50)

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))

# Graphique 1: Distribution du support des itemsets
plt.subplot(2, 2, 1)
frequent_itemsets['support'].hist(bins=20, alpha=0.7, color='blue')
plt.title('Distribution du support des itemsets')
plt.xlabel('Support')
plt.ylabel('Fréquence')

# Graphique 2: Top 10 itemsets par support
plt.subplot(2, 2, 2)
top_itemsets = frequent_itemsets.nlargest(10, 'support')
plt.barh(range(len(top_itemsets)), top_itemsets['support'])
plt.yticks(range(len(top_itemsets)), [str(list(x)) for x in top_itemsets['itemsets']], fontsize=8)
plt.title('Top 10 itemsets par support')
plt.xlabel('Support')

# Graphique 3: Relation support-confidence des règles
if len(rules) > 0:
    plt.subplot(2, 2, 3)
    plt.scatter(rules['support'], rules['confidence'], alpha=0.5, c=rules['lift'], cmap='viridis')
    plt.colorbar(label='Lift')
    plt.xlabel('Support')
    plt.ylabel('Confidence')
    plt.title('Support vs Confidence (couleur = Lift)')

# Graphique 4: Distribution du lift
if len(rules) > 0:
    plt.subplot(2, 2, 4)
    rules['lift'].hist(bins=20, alpha=0.7, color='green')
    plt.title('Distribution du lift des règles')
    plt.xlabel('Lift')
    plt.ylabel('Fréquence')

plt.tight_layout()
plt.show()

# 8. Analyse des itemsets les plus intéressants
print("\n" + "="*50)
print("ITEMSETS LES PLUS INTÉRESSANTS")
print("="*50)

# Itemsets de taille 2 avec haut support
itemsets_size2 = frequent_itemsets[frequent_itemsets['length'] == 2]
if len(itemsets_size2) > 0:
    print("Top 5 itemsets de taille 2:")
    for idx, itemset in itemsets_size2.nlargest(5, 'support').iterrows():
        print(f"  {list(itemset['itemsets'])} - Support: {itemset['support']:.3f}")

# Itemsets de taille 3
itemsets_size3 = frequent_itemsets[frequent_itemsets['length'] == 3]
if len(itemsets_size3) > 0:
    print("\nTop 5 itemsets de taille 3:")
    for idx, itemset in itemsets_size3.nlargest(5, 'support').iterrows():
        print(f"  {list(itemset['itemsets'])} - Support: {itemset['support']:.3f}")

# 9. Expérimentation avec différents paramètres
print("\n" + "="*50)
print("EXPÉRIMENTATION AVEC DIFFÉRENTS SUPPORTS")
print("="*50)

supports = [0.1, 0.15, 0.2, 0.25]
for sup in supports:
    freq_items = apriori(df_encoded, min_support=sup, use_colnames=True)
    print(f"Support {sup}: {len(freq_items)} itemsets fréquents")
    
    if len(freq_items) > 0:
        rules_tmp = association_rules(freq_items, metric="confidence", min_threshold=0.5)
        print(f"  → {len(rules_tmp)} règles (confiance ≥ 0.5)")

```

### APPLICATION DE MIN-APRIORI A WINE

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from itertools import combinations
import math

# Charger le dataset Wine et convertir en format "documents"
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

print("Dataset Wine chargé")
print(f"Shape: {df.shape}")

# 1. Conversion des données numériques en format texte pour TF-IDF
print("\n" + "="*50)
print("CONVERSION EN FORMAT TEXTE")
print("="*50)
"""
    Chaque vin devient un "document"

    Chaque caractéristique devient un "mot"

    Exemple : "alcohol_élevé malic_acid_faible ash_moyen classe_1"

"""
def create_documents_from_wine(df, feature_names):
    """
    Convertit les données numériques en documents texte
    Chaque caractéristique est discrétisée et traitée comme un "mot"
    """
    documents = []
    
    for idx, row in df.iterrows():
        doc_terms = []
        for feature in feature_names:
            value = row[feature]
            # Discrétiser la valeur en 3 catégories
            if value < df[feature].quantile(0.33):
                term = f"{feature}_faible"
            elif value < df[feature].quantile(0.66):
                term = f"{feature}_moyen"
            else:
                term = f"{feature}_élevé"
            doc_terms.append(term)
        
        # Ajouter aussi la classe comme terme
        doc_terms.append(f"classe_{int(row['target'])}")
        documents.append(" ".join(doc_terms))
    
    return documents

# Créer les documents
feature_names = ['alcohol', 'malic_acid', 'ash', 'magnesium', 'total_phenols', 'flavanoids']
documents = create_documents_from_wine(df, feature_names)

print(f"Nombre de documents: {len(documents)}")
print("\nExemple de document (premier vin):")
print(documents[0])

# 2. Calcul de la matrice TF-IDF
print("\n" + "="*50)
print("CALCUL TF-IDF")
print("="*50)
"""
Calcul des poids :

    TF (Term Frequency) : Fréquence du terme dans le document

    IDF (Inverse Document Frequency) : Importance du terme dans le corpus

    TF-IDF = TF × IDF → Poids qui mesure l'importance d'un terme

Résultat : Matrice où M[i,j] = poids TF-IDF du terme j dans le document i
"""

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
feature_names_tfidf = vectorizer.get_feature_names_out()

# Convertir en DataFrame pour plus de lisibilité
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names_tfidf)

print(f"Shape de la matrice TF-IDF: {tfidf_matrix.shape}")
print(f"Nombre de termes (mots): {len(feature_names_tfidf)}")
print("\nTop 10 termes les plus fréquents:")
print(tfidf_df.sum().sort_values(ascending=False).head(10))

# 3. Implémentation de Min-Apriori
print("\n" + "="*50)
print("IMPLÉMENTATION MIN-APRIORI")
print("="*50)

class MinApriori:
    def __init__(self, min_support=0.1):
        self.min_support = min_support
    
    def compute_support(self, itemset, tfidf_matrix):
        """
        Calcule le support Min-Apriori pour un itemset
        support(itemset) = sum over all documents of min(TF-IDF des termes dans l'itemset)
        """
        if len(itemset) == 0:
            return 0.0
        
        # Obtenir les indices des termes dans la matrice
        term_indices = [self.term_to_index[term] for term in itemset]
        """
        Pour chaque document, on prend le minimum des TF-IDF des termes

        On somme ces minimums sur tous les documents
        """
        # Pour chaque document, prendre le minimum des TF-IDF des termes de l'itemset
        min_values = []
        for doc_idx in range(tfidf_matrix.shape[0]):
            term_values = [tfidf_matrix[doc_idx, term_idx] for term_idx in term_indices]
            min_val = min(term_values) if term_values else 0.0
            min_values.append(min_val)
        
        return sum(min_values)
    
    def fit(self, tfidf_matrix, feature_names):
        self.tfidf_matrix = tfidf_matrix
        self.feature_names = feature_names
        self.term_to_index = {term: idx for idx, term in enumerate(feature_names)}
        
        # Étape 1: Itemsets de taille 1
        frequent_1_itemsets = {}
        for term in feature_names:
            itemset = frozenset([term])
            support = self.compute_support(itemset, tfidf_matrix)
            if support >= self.min_support:
                frequent_1_itemsets[itemset] = support
        
        self.frequent_itemsets = frequent_1_itemsets
        print(f"Itemsets de taille 1: {len(frequent_1_itemsets)}")
        """
        PHASE A : RECHERCHE DES ITEMSETS FRÉQUENTS
        But : Trouver tous les termes individuels avec support ≥ seuil
        """
        
        # Étape 2: Générer les itemsets de taille k
        k = 2
        while True:
            candidate_itemsets = self._generate_candidates(k)
            if not candidate_itemsets:
                break
            
            frequent_k_itemsets = {}
            for itemset in candidate_itemsets:
                support = self.compute_support(itemset, tfidf_matrix)
                if support >= self.min_support:
                    frequent_k_itemsets[itemset] = support
            
            if not frequent_k_itemsets:
                break
            
            print(f"Itemsets de taille {k}: {len(frequent_k_itemsets)}")
            self.frequent_itemsets.update(frequent_k_itemsets)
            k += 1
        
        return self
    
    def _generate_candidates(self, k):
        """Génère les candidats de taille k à partir des itemsets de taille k-1"""
        if k == 2:
            # Combinaisons des itemsets de taille 1
            items = [list(itemset)[0] for itemset in self.frequent_itemsets.keys() if len(itemset) == 1]
            return [frozenset([i, j]) for i, j in combinations(items, 2)]
        else:
            # Pour k>2, utiliser la propriété Apriori
            candidates = set()
            itemsets_k_minus_1 = [itemset for itemset in self.frequent_itemsets.keys() if len(itemset) == k-1]
            
            for i in range(len(itemsets_k_minus_1)):
                for j in range(i+1, len(itemsets_k_minus_1)):
                    itemset1 = itemsets_k_minus_1[i]
                    itemset2 = itemsets_k_minus_1[j]
                    
                    # Vérifier si les k-2 premiers éléments sont identiques
                    list1 = sorted(list(itemset1))
                    list2 = sorted(list(itemset2))
                    
                    if list1[:-1] == list2[:-1]:
                        new_itemset = frozenset(list1 + [list2[-1]])
                        candidates.add(new_itemset)
            
            return list(candidates)
    
    def get_rules(self, min_confidence=0.5):
        """Génère les règles d'association (simplifié)"""
        rules = []
        for itemset in self.frequent_itemsets.keys():
            if len(itemset) >= 2:
                itemset_list = list(itemset)
                # Générer toutes les règles possibles
                for i in range(1, len(itemset_list)):
                    for antecedent in combinations(itemset_list, i):
                        antecedent = frozenset(antecedent)
                        consequent = frozenset(itemset_list) - antecedent
                        
                        if antecedent in self.frequent_itemsets and consequent in self.frequent_itemsets:
                            confidence = self.frequent_itemsets[itemset] / self.frequent_itemsets[antecedent]
                            if confidence >= min_confidence:
                                rules.append({
                                    'antecedent': antecedent,
                                    'consequent': consequent,
                                    'support': self.frequent_itemsets[itemset],
                                    'confidence': confidence
                                })
        return rules

# 4. Application de Min-Apriori
print("\nAPPLICATION DE MIN-APRIORI")
min_apriori = MinApriori(min_support=0.05)
min_apriori.fit(tfidf_matrix, feature_names_tfidf)

print(f"\nTotal d'itemsets fréquents trouvés: {len(min_apriori.frequent_itemsets)}")

# 5. Affichage des résultats
print("\n" + "="*50)
print("RÉSULTATS MIN-APRIORI")
print("="*50)

# Itemsets de taille 1
print("\nTOP 10 ITEMSETS DE TAILLE 1:")
itemsets_1 = {k: v for k, v in min_apriori.frequent_itemsets.items() if len(k) == 1}
for itemset, support in sorted(itemsets_1.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"  {list(itemset)[0]}: support = {support:.4f}")

# Itemsets de taille 2
print("\nTOP 10 ITEMSETS DE TAILLE 2:")
itemsets_2 = {k: v for k, v in min_apriori.frequent_itemsets.items() if len(k) == 2}
for itemset, support in sorted(itemsets_2.items(), key=lambda x: x[1], reverse=True)[:10]:
    terms = list(itemset)
    print(f"  {terms[0]} + {terms[1]}: support = {support:.4f}")

# Itemsets de taille 3
print("\nITEMSETS DE TAILLE 3:")
itemsets_3 = {k: v for k, v in min_apriori.frequent_itemsets.items() if len(k) == 3}
for itemset, support in sorted(itemsets_3.items(), key=lambda x: x[1], reverse=True)[:5]:
    terms = list(itemset)
    print(f"  {terms}: support = {support:.4f}")

# 6. Génération des règles
print("\n" + "="*50)
print("RÈGLES D'ASSOCIATION")
print("="*50)

rules = min_apriori.get_rules(min_confidence=0.6)
print(f"Nombre de règles générées: {len(rules)}")

if rules:
    print("\nTOP 10 RÈGLES PAR CONFIDENCE:")
    for rule in sorted(rules, key=lambda x: x['confidence'], reverse=True)[:10]:
        ant = list(rule['antecedent'])
        cons = list(rule['consequent'])
        print(f"  {ant} → {cons}")
        print(f"    Support: {rule['support']:.4f}, Confidence: {rule['confidence']:.4f}")

# 7. Analyse des associations avec les classes
print("\n" + "="*50)
print("ASSOCIATIONS AVEC LES CLASSES DE VIN")
print("="*50)

class_terms = [term for term in feature_names_tfidf if term.startswith('classe_')]
print("Termes de classe:", class_terms)

for class_term in class_terms:
    print(f"\n--- ASSOCIATIONS POUR {class_term.upper()} ---")
    class_associations = []
    
    for itemset, support in min_apriori.frequent_itemsets.items():
        if class_term in itemset and len(itemset) > 1:
            other_terms = [term for term in itemset if term != class_term]
            class_associations.append((other_terms, support))
    
    # Trier par support
    class_associations.sort(key=lambda x: x[1], reverse=True)
    
    for terms, support in class_associations[:5]:
        print(f"  {terms} → support = {support:.4f}")

# 8. Visualisation
print("\n" + "="*50)
print("VISUALISATION")
print("="*50)

import matplotlib.pyplot as plt

# Distribution des supports
supports = list(min_apriori.frequent_itemsets.values())
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(supports, bins=20, alpha=0.7, color='lightblue', edgecolor='black')
plt.title('Distribution des supports Min-Apriori')
plt.xlabel('Support')
plt.ylabel('Fréquence')

# Top itemsets par taille
plt.subplot(1, 2, 2)
sizes = [len(itemset) for itemset in min_apriori.frequent_itemsets.keys()]
plt.hist(sizes, bins=range(1, max(sizes)+2), alpha=0.7, color='lightgreen', edgecolor='black')
plt.title('Distribution de la taille des itemsets')
plt.xlabel('Taille des itemsets')
plt.ylabel('Fréquence')

plt.tight_layout()
plt.show()
```

Apriori Classique :

Support = Fréquence d'apparition (comptage binaire)

Exemple : support({A,B}) = nombre de documents contenant A ET B

Min-Apriori : 

Support = Somme des minimums TF-IDF

Exemple : support({A,B}) = ∑ min(TF-IDF(A,d), TF-IDF(B,d))

Avantage Min-Apriori : Prend en compte l'importance des termes, pas juste leur présence/absence

Données :

    Document 1 : TF-IDF(A)=0.8, TF-IDF(B)=0.6, TF-IDF(C)=0.2

    Document 2 : TF-IDF(A)=0.3, TF-IDF(B)=0.9, TF-IDF(C)=0.7

Calcul Min-Apriori :
text

support({A,B}) = min(0.8,0.6) + min(0.3,0.9) = 0.6 + 0.3 = 0.9
support({A,C}) = min(0.8,0.2) + min(0.3,0.7) = 0.2 + 0.3 = 0.5
support({B,C}) = min(0.6,0.2) + min(0.9,0.7) = 0.2 + 0.7 = 0.9

Interprétation :

    A et B : Association forte (support=0.9)

    B et C : Association forte (support=0.9)

    A et C : Association plus faible (support=0.5)

Min-Apriori révèle les "signatures TF-IDF" des différentes classes de vin basées sur l'importance relative des caractéristiques chimiques
