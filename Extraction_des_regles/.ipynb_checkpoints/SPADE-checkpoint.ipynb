```python
#!pip install spmf



# Télécharger des datasets séquentiels populaires
def download_sequential_datasets():
    """Télécharge des datasets séquentiels depuis des sources alternatives"""
    
    datasets = {
        'kosarak': 'http://philippe-fournier-viger.com/spmf/datasets/kosarak.txt',
        'fifa': 'http://philippe-fournier-viger.com/spmf/datasets/fifa.txt',
        'BMS1': 'http://philippe-fournier-viger.com/spmf/datasets/BMS1.txt',
        'BMS2': 'http://philippe-fournier-viger.com/spmf/datasets/BMS2.txt'
    }
    
    for name, url in datasets.items():
        filename = f"{name}.txt"
        if not os.path.exists(filename):
            print(f"Téléchargement de {name}...")
            try:
                response = requests.get(url)
                with open(filename, 'w') as f:
                    f.write(response.text)
                print(f"✓ {name} téléchargé ({len(response.text)} caractères)")
            except Exception as e:
                print(f"✗ Erreur pour {name}: {e}")
        else:
            print(f"✓ {name} déjà existant")

# Télécharger les datasets
download_sequential_datasets()
```

```python
def analyze_sequential_dataset(filename, dataset_name):
    """Analyse un dataset séquentiel"""
    print(f"\n{'='*50}")
    print(f"ANALYSE DU DATASET: {dataset_name}")
    print(f"{'='*50}")
    
    sequences = []
    total_items = 0
    unique_items = set()
    
    with open(filename, 'r') as f:
        for line in f:
            line = line.strip()
            if line and not line.startswith('#'):
                # Nettoyer la ligne
                items = line.replace('-1', '').replace('-2', '').split()
                sequence = [int(x) for x in items if x]
                
                if sequence:
                    sequences.append(sequence)
                    total_items += len(sequence)
                    unique_items.update(sequence)
    
    # Calcul des statistiques
    avg_length = total_items / len(sequences) if sequences else 0
    
    print(f"Nombre de séquences: {len(sequences):,}")
    print(f"Nombre d'items uniques: {len(unique_items):,}")
    print(f"Longueur moyenne des séquences: {avg_length:.2f} items")
    #print(f"Item min/max: {min(unique_items)} / {max(unique_items)}")
    
    # Afficher quelques exemples
    print(f"\nExemples de séquences ({dataset_name}):")
    for i, seq in enumerate(sequences[:3]):
        print(f"  Séquences {i+1}: {seq}")
    
    return sequences, unique_items

# Analyser tous vos datasets
datasets = {
    'kosarak': 'kosarak.txt',
    'fifa': 'fifa.txt', 
    'BMS1': 'BMS1.txt',
    'BMS2': 'BMS2.txt'
}

results = {}
for name, filename in datasets.items():
    if os.path.exists(filename):
        sequences, items = analyze_sequential_dataset(filename, name)
        results[name] = {'sequences': sequences, 'items': items}
    else:
        print(f"\n⚠️ Fichier {filename} non trouvé pour {name}")
```

```python
# === NOUVELLE SECTION : APPEL SPADE ===
print("\n" + "="*50)
print("3.5 EXÉCUTION DE CM-SPADE")
print("="*50)

def run_spade_analysis():
    """Exécute CM-SPADE via SPMF"""
    
    # 1. Convertir le format pour SPMF
    print("📋 Conversion du format pour SPMF...")
    with open("BMS1.txt", 'r') as fin, open("BMS1_spmf.txt", 'w') as fout:
        for line in fin:
            items = line.strip().split()
            if items:
                fout.write(' '.join(items) + ' -2\n')
    
    # 2. Exécuter CM-SPADE
    print(" Exécution de CM-SPADE...")
    command = [
        'java', '-jar', 'spmf.jar', 'run', 
        'CM-SPADE', 'BMS1_spmf.txt', 'output_spmf.txt', '0.001'
    ]
    
    try:
        result = subprocess.run(command, capture_output=True, text=True, timeout=120)
        print(f" Code retour: {result.returncode}")
        
        if result.returncode == 0:
            print(" CM-SPADE exécuté avec succès!")
            if os.path.exists('output_spmf.txt'):
                file_size = os.path.getsize('output_spmf.txt')
                print(f" Fichier résultats: output_spmf.txt ({file_size} octets)")
                return True
        else:
            print(f" Erreur: {result.stderr}")
            return False
            
    except subprocess.TimeoutExpired:
        print(" Timeout - SPADE trop long")
        return False
    except Exception as e:
        print(f" Exception: {e}")
        return False

# APPEL EFFECTIF DE SPADE
success = run_spade_analysis()

if success:
    # 4. ANALYSE DES RÉSULTATS CM-SPADE (votre code existant)
    print("\n" + "="*50)
    print("4. ANALYSE DES RÉSULTATS CM-SPADE")
    print("="*50)

    def analyze_spade_results(filename):
        """Analyse des résultats de CM-SPADE"""
        print("📋 Analyse du fichier de résultats...")
        
        patterns = []
        supports = []
        
        with open(filename, 'r') as f:
            for line in f:
                line = line.strip()
                if line and '#SUP:' in line:
                    pattern_part, sup_part = line.rsplit('#SUP:', 1)
                    pattern = pattern_part.strip()
                    support = int(sup_part.strip())
                    
                    patterns.append(pattern)
                    supports.append(support)
        
        print(f"   • Nombre total de motifs trouvés: {len(patterns):,}")
        print(f"   • Support moyen: {np.mean(supports):.1f}")
        print(f"   • Support médian: {np.median(supports):.1f}")
        print(f"   • Support maximum: {np.max(supports)}")
        print(f"   • Support minimum: {np.min(supports)}")
        
        pattern_data = list(zip(patterns, supports))
        pattern_data.sort(key=lambda x: x[1], reverse=True)
        
        print("\n Top 10 des motifs les plus fréquents:")
        for i, (pattern, sup) in enumerate(pattern_data[:10], 1):
            percentage = (sup / len(sequences)) * 100
            print(f"   {i:2d}. Support: {sup:>5} ({percentage:5.2f}%) | Motif: {pattern}")
        
        return patterns, supports

    # Analyser les résultats
    if os.path.exists('output_spmf.txt'):
        patterns, supports = analyze_spade_results('output_spmf.txt')
```

### Principe Fondamental de SPADE

SPADE = Sequential PAttern Discovery using Equivalence classes

Idée clé : Découvrir les motifs séquentiels fréquents en utilisant le principe d'équivalence et le support vertical.

## ÉTAPE 0 : Préparation des Données

### Format d'entrée
- Séquence 1 : A, B, C  
- Séquence 2 : A, C  
- Séquence 3 : B, C  
- Séquence 4 : A, B, C, D  

### Format SPMF
- 1 -1 2 -1 3 -2
- 1 -1 3 -2
- 2 -1 3 -2
- 1 -1 2 -1 3 -1 4 -2

### Paramètre
- **Min Support** = 2 (au moins 2 séquences)

## ÉTAPE 1 : Base de Données Verticale
On crée la liste des positions (SID, EID) :
   Item A: [(1,1), (2,1), (4,1)]    # SID=Sequence ID, EID=Element ID
   
Item B: [(1,2), (4,2)]  

Item C: [(1,3), (2,2), (3,2), (4,3)]

Item D: [(4,4)]

   Support horizontal :

    A: 3 séquences (1,2,4)

    B: 2 séquences (1,4) ✓

    C: 4 séquences (1,2,3,4) ✓

    D: 1 séquence (4) ✗

Items fréquents (longueur 1): B, C (support ≥ 2)



## ÉTAPE 2 : Génération des Motifs de Longueur 2
Méthode : Joindre les items fréquents
2.1 Itemsets dans le même élément (simultanés)

   {B,C} : Intersection des positions de B et C
   
B: [(1,2), (4,2)]

C: [(1,3), (2,2), (3,2), (4,3)]

→ Même SID et même EID? NON → {B,C} pas simultané

2.2 Séquences d'items (l'item B puis C)

B → C : Pour chaque (SID,EID) de B, chercher C avec EID > EID
B: [(1,2), (4,2)]
- (1,2) → C dans SID=1 avec EID>2: (1,3) ✓
- (4,2) → C dans SID=4 avec EID>2: (4,3) ✓

Support(B→C) = 2 séquences (1,4) ✓

2.3 Séquences d'items (l'item C puis B)

C → B : Pour chaque (SID,EID) de C, chercher B avec EID > EID  
C: [(1,3), (2,2), (3,2), (4,3)]
- (1,3) → B dans SID=1 avec EID>3: ✗
- (2,2) → B dans SID=2 avec EID>2: ✗
- (3,2) → B dans SID=3 avec EID>2: ✗  
- (4,3) → B dans SID=4 avec EID>3: ✗

Support(C→B) = 0 ✗

Motifs longueur 2 fréquents: B→C (support=2)

### ÉTAPE 3 : Génération Récurrente (Longueur k → k+1)
Principe d'Équivalence :

Les motifs qui partent du même préfixe de longueur k-1 forment une classe d'équivalence.
3.1 À partir de B→C (support=2)

Positions de B→C: [(1,2→3), (4,2→3)]

On cherche à étendre avec les items fréquents: B, C

Extension avec B:

(B→C) → B : 
- (1,2→3) → B dans SID=1 avec EID>3: ✗
- (4,2→3) → B dans SID=4 avec EID>3: ✗
Support = 0 ✗

Extension avec C:

(B→C) → C :
- (1,2→3) → C dans SID=1 avec EID>3: ✗  
- (4,2→3) → C dans SID=4 avec EID>3: (4,4) mais C à EID=3? NON
Support = 0 ✗

3.2 Autres motifs possibles

A→B :

A: [(1,1), (2,1), (4,1)]
B: [(1,2), (4,2)]
- (1,1) → B: (1,2) ✓
- (2,1) → B: ✗  
- (4,1) → B: (4,2) ✓
Support(A→B) = 2 ✓

A→C :

A: [(1,1), (2,1), (4,1)] 
C: [(1,3), (2,2), (3,2), (4,3)]
- (1,1) → C: (1,3) ✓
- (2,1) → C: (2,2) ✓
- (4,1) → C: (4,3) ✓
Support(A→C) = 3 ✓

B→C : (déjà trouvé, support=2)
Extension de A→C:

Positions A→C: [(1,1→3), (2,1→2), (4,1→3)]

Extension avec B: (A→C)→B
- (1,1→3) → B avec EID>3: ✗
- (2,1→2) → B avec EID>2: ✗  
- (4,1→3) → B avec EID>3: ✗
Support = 0 ✗

Extension avec C: (A→C)→C  
- (1,1→3) → C avec EID>3: ✗
- (2,1→2) → C avec EID>2: ✗
- (4,1→3) → C avec EID>3: ✗
Support = 0 ✗

### ÉTAPE 4 : Résultats Finaux
Motifs Séquentiels Fréquents (support ≥ 2):
